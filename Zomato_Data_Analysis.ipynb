# Count Plot of Various Months


```python
orders["Month"] = orders["Order_Date"].dt.strftime("%B") # B - month , A - Day , Y - Year
month_counts = orders["Month"].value_counts()

plt.plot(month_counts.index, month_counts.values, color="green") 
plt.grid(color = "grey")
plt.xlabel("Months", color="blue")
plt.ylabel("Order Count", color="blue")
plt.title("Count of Orders by Month", color="red")
plt.xticks(rotation=90)
plt.show()
```

# Reading CSV

```python
customers = pd.read_csv("Customers.csv")
customers
```

# Joining orders and customers table

```python
ord_cus = pd.merge(left = orders , right = customers , on = "Customer_ID" , how = "inner")
ord_cus
```

```python
ord_cus.columns
```

# Count of Orders by various Locations

```python
location_counts = ord_cus["Customer_Location"].value_counts()

plt.bar(location_counts.index, location_counts.values, color="purple") 
plt.xlabel("Months", color="blue")
plt.ylabel("Order Count", color="blue")
plt.title("Count of Orders by various Locations", color="red")
plt.xticks(rotation=45)
plt.show()
```

```python
orders["Payment_Method"]
```

# No of times by various Payment Methods

```python
payment_counts = orders["Payment_Method"].value_counts()

plt.figure(figsize=(8, 8))
plt.pie(payment_counts.values, labels=payment_counts.index, startangle=90, autopct='%1.1f%%')
plt.title("Distribution of Orders by Payment Method")
plt.legend(loc="best")
plt.show()
```

# Number of Orders by Dish Name

```python
dish_counts = orders["Dish_Name"].value_counts()

plt.figure(figsize=(12, 6))
plt.bar(dish_counts.index, dish_counts.values, color='m', edgecolor='black')
plt.xlabel("Dish Name", color="blue")
plt.ylabel("Count", color="blue")
plt.title("Number of Orders by Dish Name", color="red")
plt.xticks(rotation=90) 
plt.show()
```
